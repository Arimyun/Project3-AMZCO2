```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axes
fig, ax = plt.subplots(figsize=(12, 6))

# Turn off axes
ax.axis('off')

# Draw arrows (xy=end point, xytext=start point)
style = "Simple,tail_width=0.6,head_width=8,head_length=8"
kw = dict(arrowstyle=style, color="black", lw=1.5)

a1= patches.FancyArrowPatch((0.75, 0.65), (0.17, 0.65), connectionstyle="arc3,rad=0.3", **kw)
a2 = patches.FancyArrowPatch((0.45, 0.65), (0.15, 0.65), connectionstyle="arc3,rad=0.3", **kw)
a3 = patches.FancyArrowPatch((0.15, 0.45), (0.45, 0.45), connectionstyle="arc3,rad=.5", **kw)
for a in [a1, a2, a3]:
    plt.gca().add_patch(a)

# Define box positions and sizes ((x,y), width, height)
box1 = patches.FancyBboxPatch((0.1, 0.5), 0.1, 0.1, boxstyle="square,pad=0.05", edgecolor='none', facecolor='aliceblue')
box2 = patches.FancyBboxPatch((0.38, 0.5), 0.15, 0.1, boxstyle="square,pad=0.05", edgecolor='none', facecolor='aliceblue')
box3 = patches.FancyBboxPatch((0.7, 0.5), 0.1, 0.1, boxstyle="square,pad=0.05", edgecolor='none', facecolor='wheat')
box4 = patches.FancyBboxPatch((0.28, 0.25), 0.04, 0.05, boxstyle="square,pad=0.05", edgecolor='none', facecolor='wheat')

# Add boxes to the plot
for box in [box1, box2, box3, box4]:
    ax.add_patch(box)

# Add text inside boxes
ax.text(0.15, 0.56, r'$\Delta$Precipitation', ha='center', va='center', fontsize=16, fontweight='bold') # Precipitation box
ax.text(0.15, 0.5, '+0.14 (6.73)', ha='center', va='center', fontsize=14, color='red') # values

ax.text(0.75, 0.56, r'$\Delta$Convergence', ha='center', va='center', fontsize=16, fontweight='bold') # Convergence box
ax.text(0.75, 0.5, '-0.11 (2.38)', ha='center', va='center', fontsize=14, color='red') # values

ax.text(0.45, 0.56, r'$\Delta$Evapotranspiration', ha='center', va='center', fontsize=16, fontweight='bold') # evapotranspiration box
ax.text(0.45, 0.5, '+0.24 (4.35)', ha='center', va='center', fontsize=14, color='red') # values

ax.text(0.3, 0.28, r'$\Delta$SM', ha='center', va='center', fontsize=16, fontweight='bold')
#ax.text(0.3, 0.24, '-3', ha='center', va='center', fontsize=14, color='red') # Temperature box

plt.tight_layout()
plt.show()
```
